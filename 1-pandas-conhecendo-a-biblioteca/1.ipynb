```python
# Busca todos os tipos de imóveis da lista comercial
dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
# permite que selecione linhas de um dataframe com base em um condição específica
# passamos uma expressão do que desejamos selecionar
# selecione apenas as linhas da nossa base de dados onde na coluna tipo existam apenas componentes da nossa lista de imoveis_comerciais
dados.query("@imoveis_comerciais in Tipo")
```

```python
dados.query("@imoveis_comerciais not in Tipo")
```

```python
df = dados.query("@imoveis_comerciais not in Tipo")
df.head()
```

```python
df.Tipo.unique()
```

```python
df_preco_tipo = df.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
df_preco_tipo.plot(kind="barh", figsize=(14,10), color="purple")
```

```python
# Percentual de cada tipo de imóvel da nossa base de dados

df.Tipo.unique()
```

```python
# para contar quantas vezes cada um desses Tipos aparecem na base de dados
df.Tipo.value_counts(normalize=True)
# normalize=True pra trazer em um formato percentual
```

```python
# converte uma series para um data frame
df.Tipo.value_counts(normalize=True).to_frame().sort_values("Tipo")
```

```python
df_percentual_tipo = df.Tipo.value_counts(normalize=True).to_frame().sort_values("Tipo")
df_percentual_tipo.plot(kind="bar", color="green", xlabel="Tipos", ylabel="Percentual")
```
